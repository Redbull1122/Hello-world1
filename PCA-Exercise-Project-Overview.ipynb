# DATA


Источник данных: https://archive.ics.uci.edu/ml/datasets/Pen-Based+Recognition+of+Handwritten+Digits

    

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
digits = pd.read_csv('../DATA/digits.csv')
```

```python
digits
```

**Cоздаlbv новый датафрейм, назовем его *pixels*, в котором будут только значения для пикселей. Для этого удалите колонку number_label.**

```python
pixels = digits.drop('number_label', axis=1)
```

```python
pixels
```

### Выводим изображение на экран

**Возьмем только один ряд пикселей - для этого возьмем первую строку в датафрейме pixels.**

```python
pixels_image = pixels.iloc[0]
```

```python
pixels_image
```

**Преобразуем полученный ранее объект Series (первая строка датафрейма) в массив numpy.**

```python
pixels_image.to_numpy()
```

**Поменяем форму этого массива numpy в размерность (8,8) с помощью reshape.**

```python
pixels_image.to_numpy().reshape(8,8)
```

**С помощью Matplotlib или Seaborn выведем этот массив на экран - должно получиться изображение цифры, написанной от руки.**

```python
sns.heatmap(data=pixels_image.to_numpy().reshape(8,8), cmap="Greens")
```

```python
sns.heatmap(data=pixels_image.to_numpy().reshape(8,8), cmap="gray", annot=True)
```

## Масштабирование данных

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_pixels = scaler.fit_transform(pixels)
```

```python
scaled_pixels
```

## Метод главных компонент (PCA - Principal Component Analysis)

**Применим метод главных компонент к промасштабированному набору пикселей, выделяя 2 главные компоненты.**

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
principal_components = pca.fit_transform(scaled_pixels)
```

**Выясним, сколько вариативности объясняется этими 2 главными компонентами.**

```python
pca.explained_variance_ratio_
```

```python
np.sum(pca.explained_variance_ratio_)
```

**Построим график scatterplot для цифр в 2-мерном пространстве главных компонент, раскрашивая график разными цветами на основе колонки number_label в исходном датафрейме.**

```python
plt.figure(figsize=(10,6),dpi=150)
plt.scatter(principal_components[:,0],principal_components[:,1], c=digits['number_label'].values)
```

**Построим "интерактивный" 3D-график с результатыми работы метода главных компонент для 3 главных компонентю**

```python
pca_model = PCA(n_components=3)
```

```python
pca_pixels = pca_model.fit_transform(scaled_pixels)
```

```python
fig = plt.figure(figsize=(12,8), dpi=150)
ax = fig.add_subplot(111, projection='3d')
ax.scatter(pca_pixels[:,0],pca_pixels[:,1],pca_pixels[:,2], c=digits['number_label']);
```
